# ColabCLI 测试 Notebook

这个 notebook 用于测试 ColabCLI 远程执行功能。

```python
# 测试 1: 基本信息
import sys
import platform
print(f"Python 版本: {sys.version}")
print(f"平台: {platform.platform()}")
print(f"处理器: {platform.processor()}")
```

```python
# 测试 2: 内存信息
import psutil
mem = psutil.virtual_memory()
print(f"总内存: {mem.total / (1024**3):.2f} GB")
print(f"可用内存: {mem.available / (1024**3):.2f} GB")
print(f"内存使用率: {mem.percent}%")
```

```python
# 测试 3: GPU 检测
try:
    import subprocess
    result = subprocess.run(['nvidia-smi', '--query-gpu=name,memory.total', '--format=csv'], 
                          capture_output=True, text=True, timeout=10)
    if result.returncode == 0:
        print("GPU 检测成功:")
        print(result.stdout)
    else:
        print("没有检测到 GPU")
except Exception as e:
    print(f"GPU 检测失败: {e}")
```

```python
# 测试 4: 变量持久化测试
test_variable = "Hello from ColabCLI!"
calculation_result = sum(range(1, 101))
print(f"测试变量: {test_variable}")
print(f"1到100的和: {calculation_result}")
```

```python
# 测试 5: 验证变量持久化
print(f"上一个单元格的变量: {test_variable}")
print(f"计算结果: {calculation_result}")
```
